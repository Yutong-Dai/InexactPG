```python
import sys
import numpy as np
from processLogFiles import *
from moralesPlot import *
from plydata import *
import pandas as pd
import warnings
```

```python
def get_datasets(loss):
    if loss == 'logit':
        datasets = ["a9a", "australian", "breast_cancer", "colon_cancer",
                "duke", "german_numer", 'heart', "ijcnn1", "ionosphere", "leu",
                "mushrooms", "phishing", "splice", "sonar", "svmguide1", "svmguide3", "w8a"
                ]
    elif loss == 'ls':
        datasets = ['abalone_scale', 'bodyfat_scale', 'cadata', 'cpusmall_scale', 'eunite2001',
                    'housing_scale', 'mg_scale', 'mpg_scale', 'pyrim_scale', 'space_ga_scale', 'blogData_train',
                    'UJIIndoorLoc', 'driftData', 'YearPredictionMSD', 'triazines_scale', 'virusShare']
    else:
        raise ValueError(f'Invalid loss:{loss}')
    return datasets
```

# LeastSquares

## last-step 1.0

```python
####### change here! ########
logdir='../IPG/test'; date = "04_06_2021"; loss = 'ls'
datasets = get_datasets(loss)
solver='naive';safeguard_opt='laststep'; safeguard_const='1.0'
sch_solver='schimdtPrescribe'; sch_safeguard_opt=''; sch_safeguard_const=''
##########
frame = summarize_data_to_csv(logdir, date, solver, loss, datasets, safeguard_opt, safeguard_const)
naive, naive_failed = prepare_morales_txt(logdir, date, solver, loss, safeguard_opt, safeguard_const, filter_condition=None, solver_ext=f'{safeguard_opt}_{safeguard_const}', overwriteFailed=True, lam_shrinks=[0.1, 0.01], percents=[0.1, 0.3, 0.5])
print("=======")

frame = summarize_data_to_csv(logdir, date, sch_solver, loss, datasets, sch_safeguard_opt, sch_safeguard_const)
# intentionally use safeguard_opt and safeguard_const to keep a copy of schimdt version for comparison
schimdt, schimdt_failed = prepare_morales_txt(logdir, date, sch_solver, loss, sch_safeguard_opt, sch_safeguard_const,
filter_condition=None, solver_ext=f'{safeguard_opt}_{safeguard_const}', overwriteFailed=True, lam_shrinks=[0.1, 0.01], percents=[0.1, 0.3, 0.5])
print("=======")
print("Selecting subsets.....")
threshold = 1
sub_frame1 = naive >> query("time >= {}".format(threshold))
sub_frame2 = schimdt >> query("time >= {}".format(threshold))
temp = np.union1d(sub_frame1['datasetid'].to_numpy(),sub_frame2['datasetid'].to_numpy())
condition  = np.sort(temp)
sub_frame1, _ = prepare_morales_txt(logdir, date, solver, loss, safeguard_opt, safeguard_const, filter_condition=condition, solver_ext=f'{safeguard_opt}_{safeguard_const}', overwriteFailed=True, lam_shrinks=[0.1, 0.01], percents=[0.1, 0.3, 0.5])
# intentionally use safeguard_opt and safeguard_const to keep a copy of schimdt version for comparison
sub_frame2, _ = prepare_morales_txt(logdir, date, sch_solver, loss, sch_safeguard_opt, sch_safeguard_const, filter_condition=condition, solver_ext=f'{safeguard_opt}_{safeguard_const}', overwriteFailed=True, lam_shrinks=[0.1, 0.01], percents=[0.1, 0.3, 0.5])

Fdiff, Fbad_case_schimdt, Fsame_case_schimdt, Fgood_case_schimdt = sanityCheck(sub_frame1, 'naive', sub_frame2, 'schimdt', metric='F', threshold=1e-6, save=False)

# idx = np.abs(sub_frame1['F'] - sub_frame2['F']) > 1e-6
# cols = ['datasetid', 'time', 'F', 'iteration']
# sub_frame1_ = sub_frame1[idx]
# sub_frame2_ = sub_frame2[idx]
# failed_check = pd.merge(sub_frame1_[cols], sub_frame2_[cols], on='datasetid', suffixes=('_naive', '_schimdt'))
# failed_check
```
